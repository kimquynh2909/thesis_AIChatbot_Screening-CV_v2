```python

from pathlib import Path
from src.preprocessing.document_parser import DocumentParser
from src.models.tfidf_matcher import TfidfMatcher
from src.models.bm25_matcher import BM25Matcher

parser = DocumentParser()
jd = parser.parse_path(Path('data/sample_job_descriptions/machine_learning_engineer.txt')).cleaned_text
resume_paths = sorted(Path('data/sample_resumes').glob('*.txt'))
resumes = [parser.parse_path(path).cleaned_text for path in resume_paths]

for model in [TfidfMatcher(), BM25Matcher()]:
    print(model.name)
    for path, score in zip(resume_paths, model.score(jd, resumes)):
        print(path.name, round(score, 4))
```
